### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
"""
En este primer análisis, debes calcular el ingreso total de cada tienda.
Sumando los valores de la columna Precio de cada conjunto de datos
de la tienda para estimar los ingresos.
"""
print("1. Ingreso total por cada tienda")


tiendas_data = {'Tienda 1': tienda,
                'Tienda 2': tienda2,
                'Tienda 3': tienda3,
                'Tienda 4': tienda4
               }

ingreso = {}
total= {}
for nombre_tienda, data in tiendas_data.items():
      total[nombre_tienda] = data['Precio'].sum()
      ingreso[nombre_tienda] = total[nombre_tienda]
      print(f"El ingreso total de {nombre_tienda} es: ${ingreso[nombre_tienda]:,.2f}")


# Gráfico de barras con las ventas por tienda
tiendas = list(ingreso.keys())
valores = list(ingreso.values())

# Crear gráfico de barras con colores
colores = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']
barras = plt.bar(tiendas, valores, color=colores)

# Añadir valores encima de cada barra
for i, valor in enumerate(valores):
    plt.text(i, valor + max(valores)*0.02, f'${valor:,.0f}',
            ha='center', va='bottom', fontweight='bold', fontsize=7)

# Formatear el eje Y con formato de moneda
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x:,.0f}'))

# Configurar etiquetas y título
plt.xlabel('Tiendas', fontweight='bold')
plt.ylabel('Ingresos ($)', fontweight='bold')
plt.title('Ingresos por Tienda', fontsize=14, fontweight='bold')

# Añadir grid para mejor lectura
plt.grid(axis='y', alpha=0.3)

# Mostrar gráfico
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
"""
En este debes calcular la cantidad de productos vendidos por categoría en cada
tienda. La idea es agrupar los datos por categoría y contar el número de ventas
de cada tipo, mostrando las categorías más populares de cada tienda.
"""
print("2. Ventas por categoría")

tiendas_data = {
                'Tienda 1': tienda,
                'Tienda 2': tienda2,
                'Tienda 3': tienda3,
                'Tienda 4': tienda4
               }

resultados = {}

for nombre_tienda, data in tiendas_data.items():
        print(f"\n {nombre_tienda.upper()}")
        print("-" * 40)


        # Contar productos por categoría
        conteo_categorias = data['Categoría del Producto'].value_counts()
        total_productos = len(data)

        # Guardar resultados
        resultados[nombre_tienda] = {
            'categorias': conteo_categorias.to_dict(),
            'total_productos': total_productos,
        }

        # Mostrar resultados
        print(f"Total de productos: {total_productos}")
        print(f"Número de categorías: {len(conteo_categorias)}")
        print("\nCategorías más populares:")

        for i, (categoria, cantidad) in enumerate(conteo_categorias.head(10).items(), 1):
            porcentaje = (cantidad / total_productos) * 100
            print(f"  {i:2d}. {categoria}: {cantidad} productos ({porcentaje:.1f}%)")

#    Obtiene las top N categorías de cada tienda
top_n = 3
print(f"\n{'='*60}")
print(f"TOP {top_n} CATEGORÍAS MÁS POPULARES POR TIENDA")
print("="*60)

top_categorias = {}

for tienda, datos in resultados.items():
    if datos['categorias']:
            # Ordenar categorías por cantidad
            categorias_ordenadas = sorted(datos['categorias'].items(),
                                        key=lambda x: x[1], reverse=True)
            top_categorias[tienda] = categorias_ordenadas[:top_n]

            print(f"\n {tienda}:")
            for i, (categoria, cantidad) in enumerate(top_categorias[tienda], 1):
                porcentaje = (cantidad / datos['total_productos']) * 100
                print(f"  {i}. {categoria}: {cantidad} ({porcentaje:.1f}%)")
    else:
            top_categorias[tienda] = []
            print(f"\n {tienda}: Sin datos de categorías")
# GRÁFICO RECOMENDADO: Subgráficos individuales para cada tienda
# Este es el más claro para comparar TOP 3 dentro de cada tienda

plt.style.use('default')  # Estilo limpio
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
axes = axes.flatten()

# Colores atractivos para cada tienda
colores = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']

for i, (tienda, top_cats) in enumerate(top_categorias.items()):
    if i < 4 and top_cats:  # Máximo 4 tiendas
        # Preparar datos
        categorias = [cat[0][:20] + '...' if len(cat[0]) > 20 else cat[0]
                     for cat in top_cats]
        cantidades = [cat[1] for cat in top_cats]

        # Crear gráfico de barras
        bars = axes[i].bar(range(len(categorias)), cantidades,
                          color=colores[i], alpha=0.8, edgecolor='white', linewidth=2)

        # Configurar el gráfico
        axes[i].set_title(f'{tienda} - TOP 3 Categorías',
                         fontsize=14, fontweight='bold', pad=20)
        axes[i].set_xlabel('Posición', fontsize=12)
        axes[i].set_ylabel('Cantidad de Productos', fontsize=12)

        # Etiquetas en el eje X
        axes[i].set_xticks(range(len(categorias)))
        axes[i].set_xticklabels([f'{j+1}°' for j in range(len(categorias))],
                               fontsize=11)

        # Valores en las barras
        for j, (bar, cantidad) in enumerate(zip(bars, cantidades)):
            axes[i].text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(cantidades)*0.01,
                        str(cantidad), ha='center', va='bottom',
                        fontweight='bold', fontsize=11)

        # Nombres de categorías debajo del gráfico
        categoria_text = '\n'.join([f'{j+1}°: {cat}' for j, cat in enumerate(categorias)])
        axes[i].text(0.02, 0.98, categoria_text, transform=axes[i].transAxes,
                    fontsize=9, verticalalignment='top',
                    bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

        # Mejorar apariencia
        axes[i].grid(axis='y', alpha=0.3, linestyle='--')
        axes[i].set_ylim(0, max(cantidades) * 1.15)

    elif i < 4:  # Si no hay datos para esta tienda
        axes[i].text(0.5, 0.5, f'{list(top_categorias.keys())[i]}\nSin datos',
                    transform=axes[i].transAxes, ha='center', va='center',
                    fontsize=14, style='italic')
        axes[i].set_xticks([])
        axes[i].set_yticks([])

# Eliminar subgráficos vacíos si hay menos de 4 tiendas
num_tiendas = len([t for t in top_categorias.items() if t[1]])
if num_tiendas < 4:
    for i in range(num_tiendas, 4):
        axes[i].remove()

# Configuración general
plt.suptitle('TOP 3 Categorías Más Populares por Tienda',
             fontsize=16, fontweight='bold', y=0.98)
plt.tight_layout()
plt.subplots_adjust(top=0.93)

# Mostrar el gráfico
plt.show()

# Información adicional
print("✅ Gráfico generado exitosamente!")
print(f"📊 Mostrando TOP 3 categorías para {num_tiendas} tiendas")
print("\n📋 Resumen:")
for tienda, top_cats in top_categorias.items():
    if top_cats:
        print(f"\n{tienda}:")
        for i, (categoria, cantidad) in enumerate(top_cats, 1):
            print(f"  {i}°: {categoria} ({cantidad} productos)")
    else:
        print(f"\n{tienda}: Sin datos")
```

# 3. Calificación promedio de la tienda


```python
'''
En este análisis debes calcular la calificación promedio de cada tienda.
'''
print("3. Calificación promedio de la tienda")

tiendas_data = {
                'Tienda 1': tienda,
                'Tienda 2': tienda2,
                'Tienda 3': tienda3,
                'Tienda 4': tienda4
               }

resultados = {}
promedio_calificacion = {}

for nombre_tienda, data in tiendas_data.items():

    #print(f"\n{nombre_tienda.upper()}")
    #print("-" * 40)

    promedio_calificacion= data['Calificación'].mean()
    resultados[nombre_tienda] = promedio_calificacion
    print(f"La calificación promedio de {nombre_tienda} es: {resultados[nombre_tienda]:.2f}")
    #print(f"El ingreso total de {nombre_tienda} es: ${ingreso[nombre_tienda]:,.2f}")

tiendas = list(resultados.keys())
calificaciones = list(resultados.values())

# Calcular porcentajes (basado en el total de todas las calificaciones)
total_calificaciones = sum(calificaciones)
porcentajes = [(cal/total_calificaciones)*100 for cal in calificaciones]

# Calcular porcentajes relativos (basado en calificación máxima posible, asumiendo escala 1-5)
max_calificacion = 5.0
porcentajes_relativos = [(cal/max_calificacion)*100 for cal in calificaciones]

# ============================================================================
# GRÁFICO 1: Gráfico Circular (Pie Chart) - Distribución Porcentual
# ============================================================================
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']

plt.figure(figsize=(10, 8))
wedges, texts, autotexts = plt.pie(porcentajes, labels=tiendas, colors=colors, autopct='%1.1f%%',
                                   startangle=90, explode=(0.05, 0.05, 0.05, 0.05))
plt.title('Distribución Porcentual de Calificaciones\n(Proporción del Total)', fontweight='bold', fontsize=14)

# Mejorar la legibilidad
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')

plt.show()
```

# 4. Productos más y menos vendidos

```python
'''
En este análisis debes identificar los productos más vendidos y los productos
menos vendidos en cada tienda.
'''
print("4. Productos más y menos vendidos")

tiendas_data = {
                'Tienda 1': tienda,
                'Tienda 2': tienda2,
                'Tienda 3': tienda3,
                'Tienda 4': tienda4
               }

resultados = {}
conteo_productos = {}

for nombre_tienda, data in tiendas_data.items():
    print(f"\n{nombre_tienda.upper()}")
    print("-" * 40)

    # Contar productos por categoría
    conteo_productos = data['Producto'].value_counts()
    total_productos = len(data)
    resultados[nombre_tienda] = {
        'productos': conteo_productos.to_dict(),
        'total_productos': total_productos,
    }

    print(f"Total de productos: {total_productos}")

    print("\nProductos más vendidos:")
    for i, (producto, cantidad) in enumerate(conteo_productos.head(10).items(), 1):
        porcentaje = (cantidad / total_productos) * 100
        print(f"  {i:2d}. {producto}: {cantidad} ({porcentaje:.1f}%)")

    print("\nProductos menos vendidos:")
    for i, (producto, cantidad) in enumerate(conteo_productos.tail(10).items(), 1):
        porcentaje = (cantidad / total_productos) * 100
        print(f"  {i:2d}. {producto}: {cantidad} ({porcentaje:.1f}%)")

```
